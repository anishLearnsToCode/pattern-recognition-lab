# Curve Fitting Using linear Regression

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

### The Hypothesis Function
$$ h_{\theta}(x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n$$

````python
def Hypothesis(X, theta):
    return X.dot(theta)
````

### The Cost Function
$$ J_{\theta}(X, Y) = \frac{1}{m} \sum (h_{\theta}(x_i) - y_i) ^ 2$$

````python
def Error(X,Y,theta):
    hx = Hypothesis(X, theta)
    return ((hx - Y) ** 2).mean()
````

### Computig the Gradient 
$$
\frac{\delta J}{\delta \theta_j} = \frac{1}{m} \Sigma_{i \in m} (h(x^{(i)}) - y^{(i)}) x^{(i)}_j
$$

Vectorized will be 

$$
\frac{\delta J}{\delta \theta} = \frac{1}{m} X^{T} \cdot (h(X) - Y)
$$

````python
def gradient(X, Y, theta):
    m, _ = X.shape
    hx = Hypothesis(X, theta)
    return X.T.dot(hx - Y) / m
````

### Performing Gradient Descent

```python
for i in numberOfiterations:
    theta := theta - learningRate * gradient
```

````python
def gradient_descent(X, Y, lr=0.1, max_itr=100):
    m, n = X.shape
    err_list = []
    theta = np.zeros((n, 1))
    
    for i in range(max_itr):
        err = Error(X, Y, theta)
        err_list.append(err)
        grad = gradient(X, Y, theta)
        # gradient Descent
        theta = theta - lr * grad
        if i % 100 == 0:
            print(f'Iteration no. {i} and cost is {err}')
    return theta, err_list
````

## Combining all parts into a Multi-variate Regression Model

````python
class LinearRegressionModel():    
    def __init__(self, lr=.1, max_itr=100):
        self.lr = lr
        self.max_itr = max_itr
    
    def fit(self, X, Y):
        X_m = np.hstack((np.ones((X.shape[0],1)), X))
        self.theta, err_list = gradient_descent(X_m, Y, lr=self.lr, max_itr=self.max_itr)
        return err_list
    
    def predict(self, X):
        X_m = np.hstack((np.ones((X.shape[0],1)), X))
        return Hypothesis(X_m, self.theta)
    
    def evaluate(self, X, Y_actual):
        Y_pred = self.predict(X)
        num = ((Y_actual - Y_pred) ** 2).sum()
        denom = ((Y_actual - Y_actual.mean()) ** 2).sum()
        return (1 - (num / denom)) * 100
````
