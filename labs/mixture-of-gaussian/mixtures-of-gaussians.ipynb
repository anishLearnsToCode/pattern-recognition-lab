```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

A _superposition_ of $K$ gaussian densities is known as a **mixture of Gaussians**. We define the probability of $x$ in a mixture of Gaussians as follows:

$$
    p(\mathbf x) = \sum_{i=1}^K \pi_k \mathcal N(\mathbf x | \boldsymbol \mu_k, \boldsymbol \Sigma_k)
$$

* $N(\mathbf x | \boldsymbol \mu_k, \boldsymbol \Sigma_k)$ is called a _component_
* $\pi_k$ is called the _mixing coeficient_

```python
x = np.linspace(-3, 3, 100)
mu1, mu2, mu3 = -1.2, 1.2, 0.6
sigma1, sigma2, sigma3 = 0.4, 0.4, 1
pi1, pi2, pi3 = 0.4, 0.2, 0.3

c1 = norm.pdf(x, loc=mu1, scale=sigma1)
c2 = norm.pdf(x, loc=mu2, scale=sigma2)
c3 = norm.pdf(x, loc=mu3, scale=sigma3)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 5))

ax[0].plot(x, c1, label="c1")
ax[0].plot(x, c2, label="c2")
ax[0].plot(x, c3, label="c3")
ax[1].plot(x, pi1 * c1 + pi2 * c2 + pi3 * c3)
ax[1].set_title('GMM')

ax[0].legend()
```
